```python
hist = pd.read_csv("Divvy_Bicycle_Stations_-_Historical_hourly.csv")
trips = pd.read_csv("divvy_trips_data_final.csv")
```

```python
trips["date"]=pd.to_datetime(trips["start_time"]).dt.date
demand=trips.value_counts(["start_station_name", "date"]).to_frame('counts')
demand.to_csv("./demand.csv")
demand.head()
```

```python
hist=hist.loc[hist["Status"]=="In Service"]
hist["date"]=pd.to_datetime(hist["Timestamp"]).dt.date
hist["Available Bikes"]=hist["Available Bikes"].astype(int)
supply = hist.groupby(["Station Name","date"])["Available Bikes"].sum().to_frame()
supply.to_csv("./supply.csv")
supply.head()
```

```python
s = pd.read_csv("supply.csv")
s=s.sort_values(by=["Station Name","date"])
d = pd.read_csv("demand.csv")
d=d.sort_values(by=["start_station_name", "date"])
```

```python
d["ratio"]=0
```

```python
for i in d.index:
    label = s.loc[(s["Station Name"]==d["start_station_name"][i]) & (s["date"]==d["date"][i])]["Available Bikes"]
    if len(label) !=0:
        d["ratio"][i]=float((label/d["counts"][i]).values)
```

```python
ratio=d.groupby(["start_station_name"]).mean()
ratio.to_csv("./ratio.csv")
```

```python
stations["Supply_to_Demand"]=0
stations["Demand"]=0
ratios = pd.read_csv("ratio.csv")
for i in stations.index:
    stations["nearest_train_station_distance"][i] = stations["nearest_train_station_distance"][i] * 0.621
    stations["nearest_bus_station_distance"][i] = stations["nearest_bus_station_distance"][i] * 0.621
    ind1=ratios.loc[ratios["start_station_name"]==stations["Station Name"][i]]["counts"]
    ind2=ratios.loc[ratios["start_station_name"]==stations["Station Name"][i]]["ratio"]
    if len(ind1)!=0 and len(ind2)!=0:
        stations["Demand"][i]=ind1.values
        stations["Supply_to_Demand"][i]=ind2.values
stations
```

```python
bus_weight = 0.5
train_weight = 0.3
dist_s1_weight = 0.8
dist_s2_weight = 0.5
dist_s3_weight = 0.3
stod_weight = 1.5

def find_efficiency(station_data):
    for i in range(len(station_data)):
        efficiency = station_data.loc[i]["nearest_bus_station_distance"] * bus_weight
        efficiency = efficiency + station_data.loc[i]["nearest_train_station_distance"] * train_weight
        

        dist = []
        for j in range(len(station_data)):
            if i != j:
                dist.append(geopy.distance.geodesic(literal_eval(station_data.loc[i]["Location"]), literal_eval(station_data.loc[j]["Location"])).miles)
        dist.sort()

        efficiency = efficiency + dist[0]*dist_s1_weight
        efficiency = efficiency + dist[1]*dist_s2_weight
        efficiency = efficiency + dist[2]*dist_s3_weight
        
        efficiency = efficiency + station_data.loc[i]["Supply_to_Demand"] * stod_weight

        station_data.loc[i,"efficiency"] = efficiency
```

```python
find_efficiency(stations)
stations
```

```python
bus_weight = 0.5
train_weight = 0.3
dist_s1_weight = 0.8
dist_s2_weight = 0.5
dist_s3_weight = 0.3
stod_weight = 1.5

def find_efficiency(station_data):
    for i in range(len(station_data)):
        efficiency = station_data.loc[i]["nearest_bus_station_distance"] * bus_weight
        efficiency = efficiency + station_data.loc[i]["nearest_train_station_distance"] * train_weight
        

        dist = []
        for j in range(len(station_data)):
            if i != j:
                dist.append(geopy.distance.geodesic(literal_eval(station_data.loc[i]["Location"]), literal_eval(station_data.loc[j]["Location"])).miles)
        dist.sort()

        efficiency = efficiency + dist[0]*dist_s1_weight
        efficiency = efficiency + dist[1]*dist_s2_weight
        efficiency = efficiency + dist[2]*dist_s3_weight
        if 0.3<station_data.loc[i]["Supply_to_Demand"]<0.5:
            efficiency = efficiency + station_data.loc[i]["Supply_to_Demand"] * stod_weight
        else:
            efficiency = efficiency - station_data.loc[i]["Supply_to_Demand"] * stod_weight


        station_data.loc[i,"efficiency"] = efficiency
```
