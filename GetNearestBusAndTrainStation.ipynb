```python
import pandas as pd
import numpy as np
```

```python
divvy_station_data = pd.read_csv("Divvy_Bicycle_Stations_-_All_-_Map.csv")

divvy_station_data.head()
```

```python
bus_stops_data = pd.read_csv("CTA_BusStops.csv")
bus_stops_data.head()
bus_stops_data_location_list = list(zip(bus_stops_data.PUBLIC_NAM,bus_stops_data.the_geom))
bus_stops_data_location_list
```

```python
tain_station_data = pd.read_csv("CTA_-_System_Information_-_List_of__L__Stops.csv")
tain_station_data.head()
tain_station_data_location_list = list(zip(tain_station_data.STATION_NAME,tain_station_data.Location))
tain_station_data_location_list
```

```python
from geopy.distance import geodesic as GD
```

### Find Nearest Bus and Train Station
1. Get Geopgraphic Cooridinates of all train and bus stations
2. Find Shortest one among all distances calculated

```python
import sys
tain_station_data_location_list
bus_stops_data_location_list

def getNearestBusStation(divvy_station_name,divvy_station_location):

    min_distance = sys.float_info.max
    nearest_station_name = ""

    for each in bus_stops_data_location_list:
        
        bus_station_location_tuple = (each[1].split(" ")[2].strip(")"),each[1].split(" ")[1].strip("("))
        divvy_station_location_tuple = (divvy_station_location.split(",")[0].strip("("),divvy_station_location.split(",")[1].strip(")"))

        distance = GD(divvy_station_location_tuple,bus_station_location_tuple).km

        if(min_distance>distance):
            min_distance = distance
            nearest_station_name = each[0]


    return nearest_station_name,min_distance

def getNearestTrainStation(divvy_station_name,divvy_station_location):

    min_distance = sys.float_info.max
    nearest_station_name = ""

    for each in tain_station_data_location_list:
        
        train_station_location_tuple = (each[1].split(",")[0].strip("(").strip(),each[1].split(",")[1].strip(")").strip())
        divvy_station_location_tuple = (divvy_station_location.split(",")[0].strip("("),divvy_station_location.split(",")[1].strip(")"))

        distance = GD(divvy_station_location_tuple,train_station_location_tuple).km

        if(min_distance>distance):
            min_distance = distance
            nearest_station_name = each[0]


    return nearest_station_name,min_distance

def getNearestBusAndTrainStation(row):
    row["nearest_bus_station_name"] ,row["nearest_bus_station_distance"] = getNearestBusStation(row["Station Name"],row["Location"])
    row["nearest_train_station_name"] ,row["nearest_train_station_distance"] = getNearestTrainStation(row["Station Name"],row["Location"])
    return row
```

```python
divvy_station_data = divvy_station_data.apply(getNearestBusAndTrainStation, axis=1)
divvy_station_data.head(100)
```
