```python
def get_map_visualization(df):
    m = folium.Map(location=[42.060895,-88.212275], tiles='OpenStreetMap', zoom_start=5)
#     df = pd.read_csv(self.sheet+"_updated.csv")
        
    markerCluster = MarkerCluster().add_to(m)
    for i, row in df.iterrows():
        lat = df.at[i,'Latitude']
        lng = df.at[i,'Longitude']

        name = df.at[i,'Station Name']
        popup = str(df.at[i,'Station Name'])

        folium.Marker(location=[lat, lng],popup=name, icon=folium.Icon(color='red')).add_to(markerCluster)

    m.save("view.html")
```

```python
get_map_visualization(df_new)
```
